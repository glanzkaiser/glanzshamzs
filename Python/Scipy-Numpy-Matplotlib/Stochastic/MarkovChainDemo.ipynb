```python
import numpy as np
import matplotlib.pyplot as plt

from markovchain import MarkovChain
```

```python
P = np.array([[0.8, 0.2], [0.1, 0.9]]) # Transition matrix
mc = MarkovChain(P, ['Rain', 'Sunny'])
mc.draw("../img/markov-chain-two-states.png")
```

```python
P = np.array([
    [0.8, 0.1, 0.1],
    [0.1, 0.7, 0.2],
    [0.1, 0.7, 0.2],
])
mc = MarkovChain(P, ['Freya', 'Lucrif', 'Hamzst'])
mc.draw("../img/markov-chain-three-states.png")
```

```python
P = np.array([
    [0.8, 0.1, 0.1, 0.0],
    [0.1, 0.7, 0.0, 0.2],
    [0.1, 0.0, 0.7, 0.2],
    [0.1, 0.0, 0.7, 0.2]
])
mc = MarkovChain(P, ['Caldraz', 'Mischkra', 'Freya', 'Zalsvik'])
mc.draw("../img/markov-chain-four-states.png")
```
